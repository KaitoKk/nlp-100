### 00. 文字列の逆順
文字列”stressed”の文字を逆に（末尾から先頭に向かって）並べた文字列を得よ

```python
str = "stressed"
reverse_str = ""

for i in range(1, len(str)+1):
    reverse_str += str[-i]

print(reverse_str)

print(str[::-1])
```

## 02. 「パトカー」＋「タクシー」＝「パタトクカシーー」
「パトカー」＋「タクシー」の文字を先頭から交互に連結して文字列「パタトクカシーー」を得よ．

```python
str02_1 = "パトカー"
str02_2 = "タクシー"
connected_str = ""

for i in range(0, len(str02_1)):
    connected_str += str02_1[i] + str02_2[i]
    
print(connected_str)
```

## 03. 円周率
“Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics.”という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成せよ．

```python
str03 = 'Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics.'
str03_list = str03.split(' ')

print([len(str.replace(',', '').replace('.', '')) for str in str03_list])
```

## 04. 元素記号
“Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.”という文を単語に分解し，1, 5, 6, 7, 8, 9, 15, 16, 19番目の単語は先頭の1文字，それ以外の単語は先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）への連想配列（辞書型もしくはマップ型）を作成せよ．

```python
str04 = 'Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.'
char_index_list = [1, 5, 6, 7, 8, 9, 15, 16, 19]
element_dict = {}

for index, word in enumerate(str04.split(' ')):
    key = word[0] if index + 1 in char_index_list else word[0:2]
    element_dict[key] = index
    
print(element_dict)
```
